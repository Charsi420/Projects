```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
import numpy as np
```

# Q1

```python
data = pd.read_csv('jsrt_metadata.csv')
#discarding data with NaN
data = data.dropna()
data
```

```python
#Histogram
plt.pyplot.figure(figsize=(12, 20))
plt.pyplot.title('Diagnosis')
plt.pyplot.ylabel('Type Of Disease')
plt.pyplot.xlabel('NUmber Of Patients')
graph=plt.pyplot.hist(data['diagnosis'],bins = 63,color='yellow',ec='blue',orientation='horizontal',histtype='bar')
```

```python
#Total Number Of Patients Of Each Diagnosis(Disease)
df = data['diagnosis'].value_counts()
df
```

```python
#Extra: Bar Graph
graph=df.plot(kind='bar',figsize=(20,10))
```

# Q:2

```python
#Separating Malignant
data2 = data[data['state'] == 'malignant']

#Number of Patients With Each Diagnosis(Disease) With Respect To Malignant
df=data2['diagnosis'].value_counts()
df
```

```python
#data for using explode
a=[]
a.append(0.1)
a.append(0.2)
for i in range(2,len(df)):
    a.append(0)

#Plotting Pie Chart Of Diagnosis With Respect To Malignant
lbl=df.index.tolist()
plt.pyplot.figure(figsize=(15, 25))
plt.pyplot.title('Diagnosis With Respect To Malignant')
graph=plt.pyplot.pie(data2['diagnosis'].value_counts(),explode=a,shadow = True,rotatelabels=True,
                     autopct='%1.2f%%',startangle=25,labels=lbl)
```

```python
#Separating Benign
data3 = data[data['state'] == 'benign']

#List of Diagnosis With Respect To Benign
df=data3['diagnosis'].value_counts()
df
```

```python
#Data for using explode
a=[]
a.append(0.1)
a.append(0.2)
for i in range(2,len(df)):
    a.append(0)
    
#Plotting Pie Chart Of Diagnosis With Respect To Benign
lbl=df.index.tolist()
plt.pyplot.figure(figsize=(15, 25))
plt.pyplot.title('Diagnosis With Respect To Benign')
graph=plt.pyplot.pie(data3['diagnosis'].value_counts(),explode=a,shadow = True,rotatelabels=True,
                     autopct='%1.2f%%',labels=lbl)
```
